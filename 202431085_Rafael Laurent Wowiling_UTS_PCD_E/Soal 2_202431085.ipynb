## Import

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Load and display the original image

```python
img2 = cv2.imread("assets/gambar-backlight.jpeg")
rgb2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
baris = rgb2.shape[0]
kolom = rgb2.shape[1]
plt.imshow(rgb2)
plt.title("Gambar asli")

def show(x,y):
    plt.imshow(x, cmap='gray')
    plt.title(y)
    plt.axis('on')
```

## Convert RGB image to grayscale 

```python
grayscale = np.zeros((baris, kolom), dtype=np.uint8)
for i in range(baris):
    for j in range(kolom):
        grayscale[i,j]=np.uint8(np.mean(rgb2[i,j]))

show(grayscale, "Grayscale")
```

## Increase grayscale image brightness

```python
beta = 40
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = grayscale[x,y] + beta
        citra_cerah[x,y] = gyx

citra_cerah = citra_cerah.astype(np.uint8)
show(citra_cerah, "Grayscale cerah")
```

## Increase grayscale image contrast

```python
alpha = 1.2
citra_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = grayscale[x,y] * alpha
        citra_kontras[x,y] = gyx

citra_kontras = citra_kontras.astype(np.uint8)
show(citra_kontras, "Grayscale kontras")
```

## Adjust brightness and contrast of grayscale image

```python
beta = 30
alpha = 1.2
citra_hasil = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = alpha * grayscale[x,y] + beta
        citra_hasil[x,y] = gyx

citra_hasil = citra_hasil.astype(np.uint8)
show(citra_hasil, "Grayscale cerah dan kontras")
```
