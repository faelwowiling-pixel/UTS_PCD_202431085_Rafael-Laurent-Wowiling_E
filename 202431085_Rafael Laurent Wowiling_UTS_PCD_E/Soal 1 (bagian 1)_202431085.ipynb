## Import

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Membaca Gambar

```python
img = cv2.imread('assets/gambar-nama.jpeg')
```

## BGR to RGB

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## Menampilkan Gambar

```python
plt.imshow(rgb)
```

## Adjust image brightness

```python
alpha = 1
beta = 30    
adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

kernel_sharpening = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
sharpened = cv2.filter2D(adjusted, -1, kernel_sharpening)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Gambar Asli')
plt.imshow(rgb)
plt.axis('off')

plt.show()
```

## Display RGB color channel detection

```python
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
plt.title('Asli')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB)[:,:,0], cmap="gray")
plt.title('Deteksi Merah')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB)[:,:,1], cmap="gray")
plt.title('Deteksi Hijau')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB)[:,:,2], cmap="gray")
plt.title('Deteksi Biru')

plt.tight_layout(pad=2.0)
plt.show()
```

```python
merah=cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB)[:,:,0]
fig, axs = plt.subplots(1,2, figsize = (15,5))
hist = cv2.calcHist([merah],[0],None,[255],[0,256])

axs[0].imshow(merah, cmap='gray')
axs[1].plot(hist)
plt.show()
```

```python
hijau=cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB)[:,:,1]
fig, axs = plt.subplots(1,2, figsize = (15,5))
hist = cv2.calcHist([hijau],[0],None,[256],[0,256])

axs[0].imshow(hijau, cmap='gray')
axs[1].plot(hist)
plt.show()
```

```python
biru=cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB)[:,:,2]
fig, axs = plt.subplots(1,2, figsize = (15,5))
hist = cv2.calcHist([biru],[0],None,[256],[0,256])

axs[0].imshow(biru, cmap='gray')
axs[1].plot(hist)
plt.show()
```
